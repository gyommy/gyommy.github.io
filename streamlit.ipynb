```python
import pandas as pd
import seaborn as sns
```

```python
link = 'https://raw.githubusercontent.com/murpi/wilddata/master/quests/cars.csv'
data = pd.read_csv(link)
```

```python
data
```

```python
data_corr = data[['mpg', 'weightlbs', 'hp', 'continent', 'cubicinches']]
```

```python
sns.pairplot(data_corr, hue="continent")
```

```python
sns.set(rc = {'figure.figsize' : (10, 5)})
sns.boxplot(data = data,
            x = 'hp',
            y = 'continent')
```

```python
sns.displot(data, x="hp", hue="continent", kind="kde", multiple="stack")
```

```python
sns.jointplot(
    data=data,
    x="hp", y="weightlbs", hue="continent"
)
```
